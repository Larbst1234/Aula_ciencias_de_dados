Tamanho dos pontos de dispersão, usaremos o seguinte comando:

```python
plt.title("Gráfico Salto vs Peso")
sns.scatterplot(x = 'Salto (cm)' , y = 'Peso (kg)', data = dados1, s=200, color = "purple")
plt.xlabel('Salto', fontsize=15)
plt.ylabel('Peso', fontsize=17)
```

Tamanho do gráfico, utilizar o comando plt.figure(figsize=(,) antes de todos os comandos.
